```python
def get_time_interval(minute):
    try:
        interval = int(minute)
        return interval
    except ValueError:
        print("Time interval must be an integer\n")
        minute_ = input("\nEnter the time interval by minutes:" )
        get_time_interval(minute_)

time_input = input("\nEnter the time interval by minutes:" )
time_interval = get_time_interval(time_input)
time_interval
```

```python
import time

from ntscraper import Nitter

def get_tweets(username):
    instances = ['nitter.esmailelbob.xyz', 'nitter.privacydev.net', 'nitter.net', 'nitter.pussthecat.org']
    for instance in instances:
        try:

            scraper = Nitter(instances=instance)

            return scraper.get_tweets(username)

        except Exception as e:

            if 'Instance has been rate limited' in str(e):

                print(f'Instance {instance} has been rate limited. Trying another instance...')

                time.sleep(60)  # wait for 60 seconds before trying another instance
            else:
                raise e
            
    raise Exception('Failed to get tweets after trying all instances')

# example usage

username = 'elonmusk'
tweets = get_tweets(username)
```

```python
import time

import threading


def spinner():
    spinner_chars = ['-', '/', '\\', '|']
    for i in range(50):
        for char in spinner_chars:
            print(f'\r{char}', end='')
            time.sleep(0.1)


def time_display():

    start_time = time.time()
    elapsed_time = 0
    while elapsed_time < 10:
        elapsed_time = time.time() - start_time
        remaining_time = 10 - elapsed_time
        print(f'\rTime remaining: {remaining_time:.1f} seconds', end='')
        time.sleep(1)




spinner_thread = threading.Thread(target=spinner)
time_thread = threading.Thread(target=time_display)

spinner_thread.start()
time_thread.start()

spinner_thread.join()
time_thread.join()
```

```python
import sys
import time

def spinner(seconds: int):
    start_time = time.time()
    elapsed_time = 0
    spinner_symbols = ['-', '\\', '|', '/']
    index = 0

    while elapsed_time < seconds:
        remaining_time = seconds - elapsed_time
        sys.stdout.write(f"\r{spinner_symbols[index]} Remaining Time: {remaining_time:.1f} seconds")
        sys.stdout.flush()
        time.sleep(0.1)
        elapsed_time = time.time() - start_time
        index = (index + 1) % len(spinner_symbols)
    sys.stdout.flush()

spinner(10)
```
